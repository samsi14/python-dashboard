```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
#to avoid encoding error, use 'unicode escape'
```

```python
df.shape
```

```python
df.head(2)
```

```python
#Data Cleaning..................
```

```python
df.info()
```

```python
#drop column,,,,, axis=1 means 1 complete vertical row
df.drop(['Status','unnamed1'], axis=1, inplace=True) 
```

```python
df.info()
```

```python
#NULL Values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head(2)
```

```python
df.info()
```

```python
#change data type
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#see columns
df.columns
```

```python
#change column name, if we add inplace=True, then it will change the column name permanently
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
#describe() :- returns description of the data in the DataFrame (i.e. count, mean, std etc)
df.describe()
```

```python
#describe for specific column
df[['Age','Orders','Amount']].describe()
```

```python
#DATA ANALYSIS....Visualization  GENDER
```

```python
df.columns
```

```python
sns.countplot(x='Gender', data=df)
```

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
          #   from above graphs we can see that most of the buyers are females and even purchasing power of females are greater than men
```

```python
#AGE
```

```python
sns.countplot(x='Age Group', data=df, hue='Gender')
```

```python
ab=sns.countplot(x='Age Group', data=df, hue='Gender')
for bars in ab.containers:
    ab.bar_label(bars)
```

```python
# amount vs age group
df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
ax=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=ax)
```

```python
# from above graphs we can see that most of the buyers are from age group 26-35 yrs females.
```

```python
#STATE....
```

```python
df.columns
```

```python
#total numbers of orders from top 10 states.
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,7)})   #set size of graph
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
#total amount from top 10 state
df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State', y='Amount', data=sales_state)
```

```python
# from above graph we can see that most of the orders and sales are from UP, MH, and karnataka resp.
```

```python
#Marital Status
```

```python
ax=sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
ax=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(4,5)})
sns.barplot(x='Marital_Status',y='Amount', data=ax, hue='Gender')
```

```python
# from above graph we can see that most of the buyers are married women and they have higher purchasing power
```

```python
#OCCUPATION
```

```python
sns.set(rc={'figure.figsize':(18,5)})
ax=sns.countplot(x='Occupation', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax=sns.barplot(x='Occupation', y='Amount', data=ax)
```

```python
#from above graph we can see that most of the workers are from IT, Healthcare and Aviation sector.
```

```python
#PRODUCT CATEGORY
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#product_category x Amount
sns.set(rc={'figure.figsize':(35,5)})
ax=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(x='Product_Category', y='Amount', data=ax)
```

```python
#from above graph we can see that most sold product are from food, clothing and electronics.
```

```python
az=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='Product_ID', y='Orders', data=az)
```

```python
* conclusion
 Married women age group 26-35 from UP, Maharashtra, and Karnataka working in IT, Healthcare and Aviation
 are more likely to buy products from Food, Clothing and Electronics category*
```
